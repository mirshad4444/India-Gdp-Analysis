```python
import os
```

```python
gsva=pd.DataFrame()
```

```python
path=("D:\Entertainment\Case Study\GDPA case study dataset\Statewise data\\")
```

```python
list_of_files=os.listdir(path)
```

```python
len(list_of_files)
```

```python
for filename in list_of_files:
    for statenames in states:
        if statenames in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[["Item","2014-15"]]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["states"]=statenames
            gsva=pd.concat([data,gsva])
```
